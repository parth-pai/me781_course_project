```python
import torch
from transformers import AutoTokenizer, AutoModel
import pandas as pd
import numpy as np
from tqdm import tqdm

prompt = "Action movie with a strong female lead"
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

# Create a function to get embeddings
def get_embedding_batch(descriptions):
    tokens = tokenizer(descriptions, return_tensors='pt', padding=True, truncation=True)
    with torch.no_grad():
        output = model(**tokens)
    return output.last_hidden_state.mean(dim=1).squeeze().detach().numpy()

embed=get_embedding_batch(prompt)
```

GRADIO CODE

```python
import gradio as gr

def get_top_movies(prompt):

    movie_df['embeddings'] = movie_df['embeddings'].str.replace('\.\.\.', ',')
    movie_df['embeddings'] = movie_df['embeddings'].apply(ast.literal_eval)

    movie_embeddings = np.array(movie_df['embeddings'].tolist())

    prompt_embedding = get_embedding_batch(prompt)
    cosine_similarities = cosine_similarity([prompt_embedding], movie_embeddings)

    top_movie_indices = np.argsort(cosine_similarities[0])[::-1][:3]

    top_movies = movie_df.iloc[top_movie_indices][['id', 'title', 'description']]
    return top_movies.to_dict(orient='records')

# Gradio Interface
iface = gr.Interface(
    fn=get_top_movies,
    inputs="text",
    outputs=gr.Table(columns=["id", "title", "description"]),
    live=True,
    interpretation="default"
)

iface.launch()
```
